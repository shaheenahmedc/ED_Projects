# Stage 6 - Track movement of SKUs between demand classes for SQL vs SP datasets

Let's first get the ADI and CV thresholds, for each demand class. From Jakob Dirken's thesis:
1. Smooth - ADI < 1.32 and CV_2 < 0.49 
2. Intermittent - ADI >= 1.32 and CV_2 < 0.49 
3. Erratic - ADI < 1.32 and CV_2 >= 0.49 
4. Lumpy - ADI >= 1.32 and CV_2 > 0.49  

But first, let's rename our current working dataframe, as it's getting a bit long:

NOTE: SLIM DON'T USE SAME CLASSIFICATIONS, THEY USE LESS. CHECK IF THEY USE ADI AND CV_2
CHECK HOW MUCH DATA THEY USE.
